```python
from PIL import Image


words = Image.open("word_matrix.png")
mask = Image.open("mask.png")
```

```python
print(words.size)
print(mask.size)
```

```python
mask = mask.resize(words.size)
```

```python
mask.putalpha(60)
words.putalpha(190)
# mask
words.paste(im=mask, box=(0,0), mask=words)

words
```
