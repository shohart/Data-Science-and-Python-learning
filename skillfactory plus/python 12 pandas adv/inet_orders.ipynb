```python
import pandas as pd

orders_df = pd.read_csv('inet_orders_data/orders.csv', sep=';')
products_df = pd.read_csv('inet_orders_data/products.csv', sep=';')

##display(orders_df.head())
##display(products_df.head())

orders_merged = products_df.merge(
    orders_df,
    left_on='Product_ID',
    right_on='ID товара',
    how='right'
    ).drop('ID товара',axis=1)
orders_merged['Total'] = orders_merged['Price'] * orders_merged['Количество']
display(orders_merged)
```

```python
orders_merged[(orders_merged['Оплачен'] == 'Да')].groupby('ID Покупателя')['Total'].sum().sort_values(ascending=False)
```
