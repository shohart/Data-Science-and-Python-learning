```python
import pandas as pd


ratings1_df = pd.read_csv('movies_data/ratings1.csv')
ratings2_df = pd.read_csv('movies_data/ratings2.csv')
##ratings1_df.head()

movies_df = pd.read_csv('movies_data/movies.csv')
##movies_df.info()

ratings1_df['userId'].nunique()

dates_df = pd.read_csv('movies_data/dates.csv')
##dates_df.info()

dates_df['date'] = pd.to_datetime(dates_df['date'])

dates_df['date'].dt.year.value_counts()[:1]
```

```python
ratings = pd.concat(
    [ratings1_df, ratings2_df],
    ignore_index=True
)
display(ratings)
```

```python
print('Число строк в таблице ratings: ', ratings.shape[0])
print('Число строк в таблице dates: ', dates_df.shape[0])
print(ratings.shape[0] == dates_df.shape[0])

display(ratings1_df.tail(1))
display(ratings2_df.head(1))

ratings = ratings.drop_duplicates(ignore_index=True)
```

```python
ratings_dates = pd.concat([ratings, dates_df], axis=1)
display(ratings_dates.tail(7))
```

```python
joined_false = ratings_dates.join(
    movies_df,
    rsuffix='_right',
    how='outer'
)
display(joined_false)
```

```python
joined = ratings_dates.join(
    movies_df.set_index('movieId'),
    on='movieId',
    how='left'
)
display(joined.head())
joined.info()
```

```python
merged = ratings_dates.merge(
    movies_df,
    on='movieId',
    how='left'
)
display(merged.head())
```

```python
print('Число строк в таблице ratings_dates: ', ratings_dates.shape[0])
print('Число строк в таблице merged: ', merged.shape[0])
print(ratings_dates.shape[0] == merged.shape[0])
```

```python
merge_ratings = ratings1_df.merge(ratings2_df, how='outer')
print('Число строк в таблице merge_ratings: ', merge_ratings.shape[0])
display(merge_ratings)
# Число строк в таблице merge_ratings: 100836
```

```python
import re 
def get_year_release(arg):
    #находим все слова по шаблону "(DDDD)"
    candidates = re.findall(r'\(\d{4}\)', arg) 
    # проверяем число вхождений
    if len(candidates) > 0:
        #если число вхождений больше 0,
	    #очищаем строку от знаков "(" и ")"
        year = candidates[0].replace('(', '')
        year = year.replace(')', '')
        return int(year)
    else:
        #если год не указан, возвращаем None
        return None
```

```python
merged['year_release'] = merged['title'].apply(get_year_release)
merged.info()
merged['movieId'].notnull().sum() - merged['year_release'].notnull().sum()
```
