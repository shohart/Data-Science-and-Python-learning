### Basic Graph

```python
from cProfile import label


x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Resize of graph
plt.figure(figsize=(6,4), dpi=150)

#plt.plot(x,y, label='2x', color='cyan', linewidth=3, marker='.', markersize=10, markeredgecolor = 'purple', linestyle = '--')
#Use a shorthand notation
#fmt = [color][marker][type]

plt.plot(x,y, 'y^--', label='2x')

## Line number 2
x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:6], x2[:6]**2, 'g', label = 'x^2')
plt.plot(x2[5:], x2[5:]**2, 'g--', label = 'x^2-forecast')
print(x2)


plt.title(' !', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.xticks([0,1,2,3,4])
#plt.yticks([0,2,4,6,8,10])
plt.legend()

plt.savefig('keithgraph.png', dpi=300)

plt.show()
```

### Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1,4,3]

patterns = ['/', 'o', '*']
colors = ['c', 'y', 'g']

bars = plt.bar(labels, values, color=colors)
plt.figure(figsize=(6,4))

for bar in bars:
    bar.set_hatch(patterns.pop(0))


# bars[0].set_hatch('/')
# bars[1].set_hatch('o')
# bars[2].set_hatch('*')



plt.show()
```
